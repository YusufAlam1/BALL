Playing around with visulaizations with the dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('injury_data.csv', encoding='latin-1')
```

```python
avg_ages = df.groupby("Likelihood_of_Injury")["Player_Age"].mean()
```

```python
avg_ages.plot(kind="bar", color=["blue", "red"], alpha=0.7)

plt.title("Avg Player_Age: Injured v Non-Injured")
plt.xlabel("Injury Likelihood (0 = No, 1 = Yes)")
plt.ylabel("Avg Age")
plt.xticks([0, 1], ["No Injury", "Injury"], rotation=0)
plt.show()
```
