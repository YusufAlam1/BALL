```python
from nba_api.stats.endpoints import (
    leaguegamefinder,
    playbyplayv2,
    playbyplayv3
)
import pandas as pd
import random
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(
    season_nullable='2024-25',
    league_id_nullable='00',
    season_type_nullable='Playoffs'
)
games = gamefinder.get_data_frames()[0]
games.head(5)
```

```python
game_ids = games['GAME_ID'].tolist()
game_ids[:5]
```

##### There are two versions of PbP Data V2 and V3. V2 Is more useful for player data, but V3 provides the additional coordinate data

```python
# game_id = random.choice(game_ids)

game_id = '0042400407'
pbp = playbyplayv2.PlayByPlayV2(game_id).get_data_frames()[0]
pbp.head().to_csv('pbp_v2.csv', index=False)
```

```python
# game_id = random.choice(game_ids)

pbp = playbyplayv3.PlayByPlayV3(game_id).get_data_frames()[0]
pbp.head()
pbp.head().to_csv('pbp_v3.csv', index=False)
```
